```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path_dataset = "/content/ESport_Earnings.csv"
```

```python
import pandas as pd

df = pd.read_csv(path_dataset, encoding='latin-1')
```

```python
df.head()
```

```python
numerical_cols = df.select_dtypes(include=['number'])  # Pilih kolom numerik saja
summary_stats = numerical_cols.describe()
print(summary_stats)
```

```python
numerical_cols = df.select_dtypes(include=['number'])  # Pilih kolom numerik saja
first_10_rows = numerical_cols.head(10)
print(first_10_rows)
```

```python
print("Jumlah record:", df.shape[0])
print("Jumlah attribute:", df.shape[1])
print("Nama attribute:", df.columns)

numerical_cols = df.select_dtypes(include=['number'])
print("Jumlah attribute dengan tipe angka:", len(numerical_cols.columns))
```

```python
categorical_cols = df.select_dtypes(include=['object'])

for col in categorical_cols:
    print(f"Jumlah label unik di {col}: {df[col].nunique()}")
```

```python
sns.histplot(data=df, x='Top_Country_Earnings', bins=30)
plt.title('Distribusi Pendapatan Pemain Esports')
plt.show()

# Box plot pendapatan per game
sns.boxplot(x='GameName', y='Top_Country_Earnings', data=df)
plt.title('Distribusi Pendapatan per Game')
plt.show()

```

## Kesimpulan Analisis Data Esports Earnings

Analisis terhadap data esports earnings mengungkapkan beberapa temuan dari data.  terdapat korelasi positif yang kuat antara jumlah jam bermain dengan total pendapatan, menunjukkan bahwa semakin banyak waktu yang dihabiskan untuk bermain, semakin besar kemungkinan untuk mendapatkan pendapatan yang lebih tinggi.



```python
corr_matrix = df.select_dtypes(include=['number']).corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Korelasi Antar Variabel Numerik')
plt.show()
```
